```python

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Пример матрицы смежности
adj_matrix = np.array([
    [0, 0, 1, 0 ,1],# eng - language
    [0, 0, 0, 1, 0],# rus - language
    [1, 0, 0, 0, 0],# USA
    [0, 1, 0, 0 ,0],# Russia
    [1, 0, 0, 0, 0] # Canada
])

# Создание графа из матрицы смежности
G = nx.from_numpy_array(adj_matrix)

# Визуализация графа
pos = nx.spring_layout(G)  # Позиционирование узлов
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=16)

# Отображение графа
plt.show()
```
